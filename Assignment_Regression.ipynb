```python
q1,q2,q3=np.percentile(df['MedInc'],[25,50,75])
iqr=q3-q1
lower_bound=q1-(1.5*iqr)
upper_bound=q3+(1.5*iqr)
print(lower_bound,upper_bound)
outlier=[]
for i in df['MedInc']:
  if i<lower_bound or i>upper_bound:
    outlier.append(i)
print(outlier)
df['MedInc']=np.where(df['MedInc']>upper_bound,upper_bound,df['MedInc'])
df['MedInc']=np.where(df['MedInc']<lower_bound,lower_bound,df['MedInc'])
```

```python
X=df.drop('Target',axis=1)
y=df['Target']
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X=scaler.fit_transform(X)
X=pd.DataFrame(X,columns=data.feature_names)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

#Linear Regression


```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
y_pred
```

```python
y_test
```

```python
slope = lr.coef_
intercept = lr.intercept_

print("Slope of best fit line is:", slope)
print("Intercept of best fit line is:", intercept)

slope_df = pd.DataFrame(slope.reshape(1, -1), columns=data.feature_names).T
slope_df.columns = ['Slope']
print(slope_df)

equation = "y = " + str(intercept)
for feature, coef in zip(data.feature_names, slope):
    equation += f" + ({coef})*{feature}"

print("Equation of best fit line:")
print(equation)
```

Linear Regression works by fitting a straight line that best describes the relationship between the input features and the target variable, minimizing the difference between predicted and actual values. It is suitable for this dataset because it provides a simple, interpretable baseline model and works well when the relationships between variables are mostly linear. Although not the most accurate, it helps establish a foundation for comparing more complex algorithms.

```python
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
print(mean_squared_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

A multiple linear regression model was fitted on the dataset to predict the target variable using all available features. The model achieved an R² score of 0.65, indicating that approximately 65% of the variance in the target variable is explained by the model.

The error metrics show that the model performs reasonably well:

Mean Squared Error (MSE): 0.44

Mean Absolute Error (MAE): 0.49

These values indicate that the model’s predictions are fairly close to the actual values on average, though there is still room for improvement. Overall, the regression model provides a good baseline performance and offers meaningful insights into the relationship between the input features and the target variable.

#KNN

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

K-Nearest Neighbors (KNN) predicts a value by looking at the ‘k’ closest data points in the feature space and averaging their target values. It works well for this dataset because it can capture nonlinear relationships without assuming any mathematical form. KNN is especially useful when similar data points tend to have similar outputs, as seen in many real-world datasets.

```python
y_new=knn.predict(X_test)
y_new

```

A K-Nearest Neighbors (KNN) regression model was evaluated using different values of k to determine the most suitable neighborhood size for prediction.

For k = 9, the model achieved:

These results indicate that k = 9 provides better predictive performance compared to k = 5, with a higher R² score and lower error values. Overall, increasing the neighborhood size to 9 improves the model's stability and accuracy for this dataset.



#Decision Tree Regressor

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor(random_state=1)
dt.fit(X_train,y_train)
```

Decision Tree Regression works by splitting the data into smaller subsets based on feature values, forming a tree-like structure where each split aims to reduce prediction error. It is suitable for this dataset because it can capture nonlinear patterns and interactions between features without requiring any scaling. However, it may overfit, which is why its performance can be less stable compared to ensemble methods.

```python
y_pred_dt=dt.predict(X_test)
y_pred_dt
```

```python
print(r2_score(y_test,y_pred_dt))
print(mean_squared_error(y_test,y_pred_dt))
print(mean_absolute_error(y_test,y_pred_dt))
```

A Decision Tree Regressor was trained and evaluated on the dataset. The model achieved an R² score of 0.5867, indicating that it explains about 59% of the variance in the target variable.

The error metrics are as follows:

Mean Squared Error (MSE): 0.5232

Mean Absolute Error (MAE): 0.4707

Compared to other models such as Linear Regression and KNN, the Decision Tree shows lower predictive accuracy and higher error values, suggesting that the tree may be overfitting the training data or failing to generalize well to unseen samples. While it captures nonlinear relationships, its overall performance is less stable and less accurate for this dataset.

##Random forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(random_state=1)
rf.fit(X_train,y_train)
```

Random Forest Regression builds multiple decision trees on different subsets of the data and averages their predictions to improve accuracy and reduce overfitting. It is suitable for this dataset because it captures complex, nonlinear relationships and handles noise effectively. Its ensemble nature makes it more stable and reliable than a single decision tree.

```python
y_pred_rf=rf.predict(X_test)
y_pred_rf
```

```python
print(r2_score(y_test,y_pred_rf))
print(mean_squared_error(y_test,y_pred_rf))
print(mean_absolute_error(y_test,y_pred_rf))
```

A Random Forest Regressor was applied to the dataset to improve prediction accuracy by combining multiple decision trees. The model performed significantly better than the previous models, achieving an R² score of 0.8080, which means it explains about 81% of the variance in the target variable.

The error metrics further show strong performance:

Mean Squared Error (MSE): 0.2430

Mean Absolute Error (MAE): 0.3261

These values indicate that the Random Forest model provides more accurate and stable predictions compared to Linear Regression, KNN, and Decision Tree models. Its ensemble nature helps reduce overfitting and captures complex relationships in the data, making it the best-performing model among those evaluated.

#Support Vector Machine

```python
from sklearn.svm import SVR
svr=SVR()
svr.fit(X_train,y_train)
```

Support Vector Regression (SVR) works by finding a curve or line that best fits the data while keeping errors within a small margin, using kernels to model complex nonlinear relationships. It is suitable for this dataset because it handles nonlinear patterns well and is robust to outliers. SVR often performs strongly when the data has a mix of linear and nonlinear trends.

```python
y_pred_svr=svr.predict(X_test)
y_pred_svr
```

```python
print(r2_score(y_test,y_pred_svr))
print(mean_squared_error(y_test,y_pred_svr))
print(mean_absolute_error(y_test,y_pred_svr))
```

A Support Vector Regression (SVR) model was trained to capture nonlinear relationships in the dataset. The model achieved a strong R² score of 0.7712, meaning it explains approximately 77% of the variance in the target variable.

The error metrics are as follows:

Mean Squared Error (MSE): 0.2896

Mean Absolute Error (MAE): 0.3648

These results indicate that the SVR model performs well, offering better accuracy than Decision Tree and Linear Regression models. However, its performance is slightly lower than that of the Random Forest model, which remains the best among the tested algorithms. Overall, SVR provides a good balance between generalization and capturing complex patterns in the data.

# Gradient Booster

```python
from sklearn.ensemble import GradientBoostingRegressor
gb=GradientBoostingRegressor()
gb.fit(X_train,y_train)
```

Gradient Boosting Regression builds a series of weak decision trees, where each new tree learns from the errors of the previous ones to gradually improve prediction accuracy. It is suitable for this dataset because it captures complex nonlinear patterns and reduces bias effectively. This sequential learning approach often leads to strong, high-accuracy performance on tabular data.

```python
y_pred_gb=gb.predict(X_test)
y_pred_gb
```

```python
print(r2_score(y_test,y_pred_gb))
print(mean_squared_error(y_test,y_pred_gb))
print(mean_absolute_error(y_test,y_pred_gb))
```

A Gradient Boosting Regressor was trained to improve prediction accuracy by combining multiple weak learners in a sequential manner. The model achieved a strong R² score of 0.7832, indicating that it explains about 78% of the variance in the target variable.

The error metrics are as follows:

Mean Squared Error (MSE): 0.2744

Mean Absolute Error (MAE): 0.3651

These results show that Gradient Boosting performs better than Linear Regression, Decision Tree, and KNN models, and slightly better than SVR. While its performance is strong, it still does not surpass the Random Forest Regressor, which remains the best-performing model in this project. Overall, Gradient Boosting provides accurate predictions and captures complex relationships effectively
